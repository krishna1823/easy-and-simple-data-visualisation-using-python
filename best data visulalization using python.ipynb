```python
# import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# read your data from working directory
data= pd.read_csv("C:/Users/lenovo/Desktop/myproject/files/Housing.csv")
```

```python
# view first few rows and all columns of data
data.head()
```

```python
# barplots for categorical variables
plt.figure(figsize=(18,12))
plt.subplot(3,3,1)
sns.countplot(data= data, x="mainroad")
plt.subplot(3,3,2)
sns.countplot(data= data, x="guestroom")
plt.subplot(3,3,3)
sns.countplot(data= data, x="basement")
plt.subplot(3,3,4)
sns.countplot(data= data, x="hotwaterheating")
plt.subplot(3,3,5)
sns.countplot(data= data, x="airconditioning")
plt.subplot(3,3,6)
sns.countplot(data= data, x="prefarea")
plt.subplot(3,3,7)
sns.countplot(data= data, x="furnishingstatus")
plt.show()
```

```python
# count categorical values 
print(data["mainroad"].value_counts())
print(data["guestroom"].value_counts())
print(data["basement"].value_counts())
```

```python
# visualize how the numerical values are distributed
plt.figure(figsize= (20,5))
plt.subplot(1,2,1)
sns.distplot(data["price"],color=  "blue")
plt.subplot(1,2,2)
sns.distplot(data["area"],color=  "red")
plt.show()
```

```python
# dot plots between two numercal variables
sns.scatterplot(data["area"], data["price"], color= "red" )
plt.show()
```

```python
sns.jointplot(data["area"], data["price"] )
plt.show()
```

```python
sns.jointplot(data["area"], data["price"], kind= "hex" )
plt.show()
```

```python
sns.jointplot(data["area"], data["price"], kind= "kde" )
plt.show()
```

```python
sns.jointplot(data["area"], data["price"], kind= "resid" )
plt.show()
```

```python
plt.figure(figsize=(15,10))
plt.subplot(2,2,1)
sns.countplot(data["bedrooms"])
plt.subplot(2,2,2)
sns.countplot(data["bathrooms"])
plt.subplot(2,2,3)
sns.countplot(data["stories"])
plt.subplot(2,2,4)
sns.countplot(data["parking"])
plt.show()
```

```python
plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
sns.swarmplot(data["mainroad"], data["price"])
plt.subplot(1,2,2)
sns.swarmplot(data["mainroad"], data["area"])
plt.show()
```

```python
plt.figure(figsize=(19,12))
plt.subplot(2,3,1)
sns.boxplot(data["mainroad"], data["area"])
plt.subplot(2,3,2)
sns.boxplot(data["guestroom"], data["area"])
plt.subplot(2,3,3)
sns.boxplot(data["basement"], data["area"])
plt.subplot(2,3,4)
sns.boxplot(data["hotwaterheating"], data["area"])
plt.subplot(2,3,5)
sns.boxplot(data["airconditioning"], data["area"])
plt.subplot(2,3,6)
sns.boxplot(data["prefarea"], data["area"])
plt.show()
```

```python
plt.figure(figsize=(15,8))
plt.subplot(2,3,1)
sns.boxplot(data["mainroad"], data["price"])
plt.subplot(2,3,2)
sns.boxplot(data["guestroom"], data["price"])
plt.subplot(2,3,3)
sns.boxplot(data["basement"], data["price"])
plt.subplot(2,3,4)
sns.boxplot(data["hotwaterheating"], data["price"])
plt.subplot(2,3,5)
sns.boxplot(data["airconditioning"], data["price"])
plt.subplot(2,3,6)
sns.boxplot(data["prefarea"], data["price"])
plt.show()
```

```python
# realtion plot between two variables
sns.relplot(x="price", y="area", data=data, hue= "mainroad")
plt.show()
```

```python
sns.relplot(x="price", y="area", data=data, hue= "prefarea")
plt.show()
```

```python
sns.relplot(x="price", y="area", data=data, hue= "parking")
plt.show()
```
